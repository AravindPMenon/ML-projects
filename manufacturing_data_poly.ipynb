### SCATTERPLOT

```python
x = df['Temperature_(°C)']
y = df['Quality_Rating']

# Fit a polynomial of degree 2 (quadratic)
coeffs = np.polyfit(x, y, deg=2)  # Get polynomial coefficients
poly_eq = np.poly1d(coeffs)  # Create a polynomial equation

# Generate values for the fitted polynomial curve
x_sorted = np.sort(x)
y_fitted = poly_eq(x_sorted)

# Plot the scatterplot and polynomial curve
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.6, label='Data Points')
plt.plot(x_sorted, y_fitted, color='red', linewidth=2, label=f'Polynomial Fit: {coeffs[0]:.3e}x^2 + {coeffs[1]:.3e}x + {coeffs[2]:.3e}')

# Customize the plot
plt.title('Polynomial Scatterplot: Temperature vs. Quality Rating', fontsize=14)
plt.xlabel('Temperature (°C)', fontsize=12)
plt.ylabel('Quality Rating', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

# IMPORTING LIBRARIES AND DATASET

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

# Extracting independent (x) and dependent (y) variables

```python

x = df['Temperature_(°C)'].values.reshape(-1, 1)
y = df['Quality_Rating'].values
```

# Create polynomial features (degree 2)

```python
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
```


# Fit the polynomial regression model

```python

model = LinearRegression()
model.fit(x_poly, y)
```

# Predicting values using the model

```python

y_pred = model.predict(x_poly)
```

# Evaluating the model

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
```

# Printing evaluation metrics

```python
print(f"Mean Squared Error: {mse:.3f}")
print(f"R-squared: {r2:.3f}")
```
